## Visualización con Python: Matplotlib

**Matplotlib** es una biblioteca para la generación de gráficos en dos dimensiones, a partir de datos contenidos en listas o arrays en el lenguaje de programación Python. Proporciona una API, pylab, diseñada para recordar a la de MATLAB.

```python
import matplotlib as plt
```

```python
help(plt)
```

Por tanto es recomendable importar matplotlib de la siguiente forma:

```python
import matplotlib.pyplot as plt
```

Es importante comenzar con el concepto de **figura**
en el que se colocarán una o mas gráficas. En matplotlib las gráficas se crean sobre un conjunto de **ejes**. Los ejes tendrán elementos adicionales como títulos de la gráfica, marcas (_ticks_), etc. Del concepto gráfica dependen cosas como el color de la gráfica, el ancho de línea usada, etc.

Por lo tanto, para crear una gráfica, es necesario crear un espacio donde se encuentre, una figura, y al menos un conjunto de ejes. Solo cuando tenemos ambos objetos creados, es posible crear la gráfica.

### La función plot

Hay diferentes formas de crear una figura y un conjunto de ejes. Una forma es por por medio de la función `matplotlib.pyplot.plot` crea de forma automática una figura y un conjunto de ejes, además que muestra la gráfica correspondiente a los datos que se le han pasado como argumentos.

```python
plt.plot([2,6, 7,3,8])
plt.show()
```

```python
plt.plot([0.1, 0.2, 0.3, 0.4],[-3, 7, 2, 8 ])
plt.show()
```

La función `matplotlib.pyplot.show` muestra la gráfica creada.

### La función subplots

Una segunda forma para crear una figura y uno o más conjuntos de ejes es por medio del uso de la función `matplotlib.pyplot.subplots`

```python
fig, ax = plt.subplots()
ax.plot([2,6,7,3,8])
plt.show()
```

### Las funciones figure y axes

Otra tercer forma de crear gráficas es por medio de las funciones `matplotlib.pyplot.figure` y `matplotlib.pyplot.axes` las cuales pueden utilizarse por separado para crear de forma explícita una figura y un conjunto de ejes. La última figura creada o referenciada sera la _figura actual_.

```python
fig = plt.figure()
ax = plt.axes()
ax.plot([2,6,7,3,8])
plt.show()
```

La interfaz **Orientada a Objetos** permite referenciar la figura y los ejes mediante variables (con métodos y atributos) que permitirán personalizar la visualizaión.

Por defecto `subplots()` crea un solo conjunto de ejes, sin embargo se puede especificar otro conjunto de ejes adicional.. Para crear dos conjuntos de ejes, hacemos:

```python
fig, ax = plt.subplots(1,2)
ax[0].plot([2,6,7,3,8])
ax[1].plot([3,4,5,2,9])
plt.show()
```

La mayor parte de los métodos que nos permiten fijar un valor a un atributo comienza por **set_**. Estos métodos suelen tener un método equivalente que comienza habitualmente por **get_** que sirve para leer el valor de dicho atributo.

# Estilos y objetos

Se graficará la función $y = xe^{-x^2}$

```python
def f(x):
    return x* np.exp(-x **2)

x = np.linspace(-5, 5, num=100)
```

```python
plt.plot(x,f(x))
```

La interfaz de `pyplot` permite agregar características al grafico por medio de la llamada a los objetos definidos para cada tipo de gráfico.

```python
plt.plot(x,f(x), 'ob', label='Función f(x)')
plt.xlabel('Eje x')
plt.ylabel('$f(x) = xe^{-x^2}$')
plt.grid(True)
plt.legend()
plt.title('$xe^{-x^2}$')
```

### Objetos de pyplot

```python
x
```

```python
f(x)
```

```python
fig, axes = plt.subplots()
axes.plot(x, f(x), 'ro', label ='Funcion f(x)')

axes.set_xlim(-2,4)
axes.set_ylim(-1,1)
axes.grid(True)
fig.savefig("grafica1.pdf")
```

```python
fig, axes = plt.subplots(1,2, sharey=True)
axes[0].plot(x,f(x), color = '#90A4AE')
axes[0].set_xlabel('Eje x izquierda')
axes[0].grid(True)
axes[1].plot(x,f(x) + 1,'#677e5d')
axes[1].set_xlabel('Eje x derecha')
axes[1].grid(False)
```

### Graficosde contorno

Se desea representar la función $g(x,y)= cos(x) + sin^2(y)$

```python
x = np.linspace(-2,2)
y = np.linspace(-2,2)

xx,yy = np.meshgrid(x,y)
```

```python
def g(x,y):
    return np.cos(x) + np.sin(y)**2
```

```python
zz = g(xx,yy)
```

```python
fig , axes = plt.subplots()
axes.contour(xx, yy, zz, cmap=plt.cm.autumn)
```
